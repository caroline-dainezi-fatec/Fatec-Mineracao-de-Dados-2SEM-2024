# 01 - Criação e Manipulação de Arrays

Crie um array Numpy com 10 números inteiros aleatórios entre 1 e 20 e extraia os elementos pares.

```python
import numpy as np

numbers = np.random.randint(1, 21, size=10)
evens = numbers[numbers % 2 == 0]

print("Original:", numbers)
print("Pares:   ", evens)
```

# 02 - Operações Aritméticas com Arrays

Crie dois arrays de 10 números inteiros aleatórios entre 0 e 50 e realize operações de soma, subtração, multiplicação e divisão elemento a elemento.

```python
numbers1 = np.random.randint(0, 51, size=10)
numbers2 = np.random.randint(0, 51, size=10)

sum = numbers1 + numbers2
subtraction = numbers1 - numbers2
multiplication = numbers1 * numbers2

division = np.divide(numbers1, numbers2, out=np.zeros_like(numbers1, dtype=float), where=numbers2!=0)

print("Numbers1:     ", numbers1)
print("Numbers2:     ", numbers2)
print("Soma:         ", sum)
print("Subtração:    ", subtraction)
print("Multiplicação:", multiplication)
print("Divisão:      ", division)
```

# 03 - Indexação e Slicing

Dado um array Numpy com 20 números aleatórios entre 10 e 100, extraia os elementos entre a 5ª e a 15ª posição.

```python
numbers = np.random.randint(10, 101, size=20)
interval = numbers[4:15]

print("Original:     ", numbers)
print("Posições 5-15:", interval)
```

# 04 - Manipulação de Matrizes

Crie uma matriz 3x3 com valores aleatórios entre 0 e 1, e encontre o valor máximo e mínimo.

```python
matrix = np.random.rand(3, 3)

max_value = np.max(matrix)
min_value = np.min(matrix)

print("Matriz 3x3:")
print(matrix)
print("\nValor máximo:", max_value)
print("Valor mínimo:", min_value)
```

# 06 - Criação de DataFrames

Crie um DataFrame com dados de 5 pessoas, onde cada pessoa tem nome, idade (entre 20 e 40), e cidade (escolha 3 cidades aleatórias).

```python
import pandas as pd

cities = ['Mauá', 'Santo André', 'Diadema']
names = ['Ana', 'Bryan', 'Carol', 'Dionísio', 'Eduarda']
ages = np.random.randint(20, 41, size=5)

random_cities = np.random.choice(cities, size=5)

df = pd.DataFrame({
    'Nome': names,
    'Idade': ages,
    'Cidade': random_cities
})

print("DataFrame:")
print(df)
```

# 07 - Criação de DataFrame com Dados Sintéticos

Crie um DataFrame fictício de vendas contendo colunas como Produto, Quantidade (entre 1 e 20), e Preço (entre 10 e 100).

```python
products = ['Maçã', 'Banana', 'Abacate', 'Uva', 'Pera']
amount = np.random.randint(1, 21, size=10)
price = np.random.uniform(10, 101, size=10)

random_products = np.random.choice(products, size=10)

df_sales = pd.DataFrame({
    'Produto': random_products,
    'Quantidade': amount,
    'Preço': price
})

print("DataFrame de Vendas:")
print(df_sales)
```

# 08 - Filtragem de Dados

Filtre o DataFrame criado no exercício anterior para mostrar apenas as vendas onde a quantidade é maior que 10.

```python
df_filtered = df_sales[df_sales['Quantidade'] > 10]

print("DataFrame Filtrado:")
print(df_filtered)
```

# 09 - Operações em Colunas

Adicione uma coluna ao DataFrame com o valor total da venda (Quantidade * Preço).

```python
df_sales['Valor Total'] = df_sales['Quantidade'] * df_sales['Preço']

print("DataFrame de Vendas com Valor Total:")
print(df_sales)
```

# 10 - Agrupamento e Agregação

Agrupe o DataFrame pelo campo Produto e calcule a soma das quantidades vendidas para cada produto.

```python
df_grouped = df_sales.groupby('Produto')['Quantidade'].sum().reset_index()

print("DataFrame Agrupado por Produto com Soma das Quantidades:")
print(df_grouped)
```

# 11 - Gráfico de Linhas

Crie um gráfico de linha que mostra a variação de preços de um produto fictício ao longo de 12 meses.

```python
import matplotlib.pyplot as plt

months = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
prices = np.random.uniform(10, 100, size=12)

df_prices = pd.DataFrame({
    'Mês': months,
    'Preço': prices
})

plt.figure(figsize=(10, 6))
plt.plot(df_prices['Mês'], df_prices['Preço'], marker='o', linestyle='-', color='#04b568')
plt.grid(True)

plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

# 16 - Gráfico de Barras com Seaborn

Use Seaborn para criar um gráfico de barras com os mesmos dados fictícios de quantidade vendida por produto.

```python
import seaborn as sns

products = ['Maçã', 'Banana', 'Abacate', 'Uva', 'Pera']
amount = np.random.randint(1, 21, size=5)

df_sales = pd.DataFrame({
    'Produto': products,
    'Quantidade': amount
})

sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
ax = sns.barplot(x='Produto', y='Quantidade', data=df_sales, palette='rocket', hue='Produto', legend=False)

plt.title('Quantidade Vendida de Produtos Fictícios')
plt.xlabel('Produto')
plt.ylabel('Quantidade Vendida')
plt.legend(['Quantidade Vendida'], loc='upper right')

plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

# 17 - Gráfico de Dispersão com Seaborn

Recrie o gráfico de dispersão da relação entre altura e peso usando Seaborn, incluindo uma linha de regressão.

```python
n_people = 100
height = np.random.normal(loc=170, scale=10, size=n_people)
weight = np.random.normal(loc=70, scale=15, size=n_people)

df = pd.DataFrame({
    'Altura': height,
    'Peso': weight
})

sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
ax = sns.regplot(x='Altura', y='Peso', data=df, scatter_kws={'color':'#7811bd', 'alpha':0.7}, line_kws={'color':'#fc9803'})

plt.title('Relação de altura x peso com linha de regressão')
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')

plt.tight_layout()
plt.show()
```

# 19 - Heatmap

Crie um heatmap para mostrar a correlação entre diferentes variáveis em um DataFrame fictício.

```python
sns.set(style="white")

n_samples = 100
df = pd.DataFrame({
    'X': np.random.normal(loc=50, scale=10, size=n_samples),
    'Y': np.random.normal(loc=30, scale=5, size=n_samples),
    'Z': np.random.normal(loc=70, scale=15, size=n_samples),
    'K': np.random.normal(loc=40, scale=20, size=n_samples)
})

corr = df.corr()

plt.figure(figsize=(10, 8))
ax = sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1, center=0)

plt.title('Heatmap de correlação de variáveis')

plt.tight_layout()
plt.show()
```
